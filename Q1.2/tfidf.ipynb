```python
def computeTF(wordDict, bow):
    tfDict = {}
    bowCount = len(bow)
    for word, count in wordDict.items():
        tfDict[word] = count/float(bowCount)
    return tfDict
```

```python
def computeIDF(docList):
    import math
    idfDict = {}
    N = len(docList)
    
    idfDict = dict.fromkeys(docList[0].keys(), 0)
    for doc in docList:
        for word, val in doc.items():
            if val > 0:
                idfDict[word] += 1
    
    for word, val in idfDict.items():
        idfDict[word] = math.log10(N / float(val))
        
    return idfDict
```

```python
def computeTFIDF(tfBow, idfs):
    tfidf = {}
    for word, val in tfBow.items():
        tfidf[word] = val*idfs[word]
    return tfidf
```

```python
result = []
for i in range(len(setdoc)):
  tfX = computeTF(setdoc[i], bow[i])


  idfs = computeIDF(setdoc)


  tfidf = computeTFIDF(tfX, idfs)
  result += ["page : " + str(i+1),sorted(tfidf.items(), key=lambda x: x[1])[-1]]
display(result)
```
